```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
from volume_density_DESI import load_z
DATA_DIR = Path("****")# Path settings
```

```python
csv_path = DATA_DIR / "nz_lightcone_gfibcut.csv"#compute_nz_lightcone.py
df = pd.read_csv(csv_path)
plt.figure(figsize=(6,4))
plt.plot(
        df["z_center"],
        df["n_VLO[deg^-2]"],
        ls="-", lw=1.5, color="red",
        label="VLO",
    )
plt.plot(
        df["z_center"],
        df["n_LOP[deg^-2]"],
        color="purple",
    label="LOP"
        )
plt.xlabel("z")
plt.ylabel("$N$[deg$^{-2}$] per $\Delta z$=0.05")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.title("Lightcone galaxy number density n(z)")
plt.legend()
plt.show()
```

```python
df_desi = pd.read_csv(DATA_DIR / "main-800coaddefftime1200-nz-zenodo.ecsv", comment="#", sep="\s+")
z_desi = load_z(DATA_DIR)
```

```python
plt.figure(figsize=(6,4))

plt.plot(z_desi, df_desi["ELG_VLO_SOUTH_DECALS"], lw=2, label="South (DECaLS)")
plt.plot(z_desi, df_desi["ELG_VLO_NORTH"],        lw=2, label="North")
plt.plot(z_desi, df_desi["ELG_VLO_SOUTH_DES"],    lw=2, label="South (DES)")
plt.plot(
        df["z_center"],
        df["n_VLO[deg^-2]"],
        ls="-", lw=1.5, color="red",
        label=f"Mock 1048 deg$^2$"
    )

plt.xlabel("z")
plt.ylabel("$N$[deg$^{-2}$] per $\Delta z$=0.05")
plt.title("VLO lightcone and DESI")
plt.legend(frameon=False, fontsize=8, ncol=2)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.plot(z_desi, df_desi["ELG_LOP_SOUTH_DECALS"], lw=2, label="South (DECaLS)")
plt.plot(z_desi, df_desi["ELG_LOP_NORTH"],        lw=2, label="North")
plt.plot(z_desi, df_desi["ELG_LOP_SOUTH_DES"],    lw=2, label="South (DES)")
plt.plot(
        df["z_center"],
        df["n_LOP[deg^-2]"],
        color="purple",
        label=f"Mock 1048 deg$^2$"
        )

plt.xlabel("z")
plt.ylabel("$N$[deg$^{-2}$] per $\Delta z$=0.05")
plt.title("LOP lightcone and DESI")
plt.ylim(top=75)
plt.legend(frameon=False, fontsize=8, ncol=2)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
