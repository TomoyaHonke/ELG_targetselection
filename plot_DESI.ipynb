```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from volume_density_DESI import ang_to_vol_density, load_z
DATA_DIR = Path("****")# Path settings
```

## Angular number density

```python
df = pd.read_csv(DATA_DIR / "main-800coaddefftime1200-nz-zenodo.ecsv", comment="#", sep="\s+")
z = load_z(DATA_DIR)
```

```python
plt.figure(figsize=(6,4))
plt.plot(z, df["ELG_VLO_SOUTH_DECALS"], label="South (DECaLS)", lw=2)
plt.plot(z, df["ELG_VLO_NORTH"], label="North", lw=2)
plt.plot(z, df["ELG_VLO_SOUTH_DES"], label="South (DES)", lw=2)
plt.xlabel("z")
plt.ylabel("$N$[deg$^{-2}$] per $\Delta z$=0.05")
plt.title("ELG_VLO — 800 coadd, efftime=1200")
plt.legend(frameon=False)
plt.grid(alpha=0.3)
plt.ylim(top=120)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(z, df["ELG_LOP_SOUTH_DECALS"], label="South (DECaLS)", lw=2)
plt.plot(z, df["ELG_LOP_NORTH"], label="North", lw=2)
plt.plot(z, df["ELG_LOP_SOUTH_DES"], label="South (DES)", lw=2)
plt.xlabel("z")
plt.ylabel("$N$[deg$^{-2}$] per $\Delta z$=0.05")
plt.title("ELG_LOP — 800 coadd, efftime=1200")
plt.legend(frameon=False)
plt.grid(alpha=0.3)
plt.ylim(top=120)
plt.tight_layout()
plt.show()
```

## comoving volume density

```python
vlo_cols   = ["ELG_VLO_SOUTH_DECALS", "ELG_VLO_NORTH", "ELG_VLO_SOUTH_DES"]
vlo_labels = ["South (DECaLS)", "North", "South (DES)"]

plt.figure(figsize=(6,4))
for col, lab in zip(vlo_cols, vlo_labels):
    n_vol = ang_to_vol_density(col, DATA_DIR)
    plt.plot(z, n_vol, lw=2, label=lab)

plt.xlabel("z")
plt.ylabel(r"$dN/dV$  [$h^3\,{\rm Mpc}^{-3}$]")
plt.title("DESI ELG_VLO — comoving number density")
plt.legend(frameon=False)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
lop_cols   = ["ELG_LOP_SOUTH_DECALS", "ELG_LOP_NORTH", "ELG_LOP_SOUTH_DES"]
lop_labels = ["South (DECaLS)", "North", "South (DES)"]

plt.figure(figsize=(6,4))
for col, lab in zip(lop_cols, lop_labels):
    n_vol = ang_to_vol_density(col, DATA_DIR)
    plt.plot(z, n_vol, lw=2, label=lab)

plt.xlabel("z")
plt.ylabel(r"$dN/dV$  [$h^3\,{\rm Mpc}^{-3}$]")
plt.title(" DESI ELG_LOP — comoving number density")
plt.legend(frameon=False)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
